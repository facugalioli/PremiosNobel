## Tema de Investigacion: Premios Nobel (1901-2019)

#### *Historia de los Datos*:  
*El Premio Nobel es un conjunto de premios internacionales anuales otorgados en varias categorías por instituciones suecas y noruegas en reconocimiento a los avances académicos, culturales o científicos*.  

*El testamento del químico, ingeniero e industrial sueco Alfred Nobel estableció los cinco premios Nobel en 1895. Los premios de Química, Literatura, Paz, Física y Fisiología o Medicina se otorgaron por primera vez en 1901. Los premios son ampliamente considerados como los más importantes. prestigiosos premios disponibles en sus respectivos campos*.  

*Entre 1901 y 2020, los Premios Nobel y el Premio de Ciencias Económicas fueron otorgados a 949 personas. El Premio Nobel es un premio internacional administrado por la Fundación Nobel en Estocolmo, Suecia, y basado en la fortuna de Alfred Nobel, inventor y empresario sueco. En 1968, Sveriges Riksbank estableció el Premio Sveriges Riksbank de Ciencias Económicas en memoria de Alfred Nobel, fundador del Premio Nobel. Cada Premio consiste en una medalla, un diploma personal y un premio en efectivo*.

## **Porcentaje de mujeres y hombres ganaron el premio Nobel desde el 1901 al 2019.**

```python
plt.pie(dfnobel1['genero'].value_counts(), 
labels=dfnobel1['genero'].value_counts().index,counterclock=False,autopct='%.2f')
plt.title('Ganadores por Genero');
```

Tenemos como resultado que desde el **1901** al **2020**, de la totalidad de los ganadores, solo un **5.85%** de los premios han sido otorgados a mujeres. Esto refleja una brecha de género en la ciencia y en otras áreas en las que se otorgan premios Nobel.

Hay muchas razones por las que esta brecha existe, incluyendo discriminación sistémica y prejuicios culturales. Durante mucho tiempo, las mujeres han enfrentado barreras para ingresar y progresar en campos científicos y tecnológicos, incluyendo la falta de oportunidades de educación y de carrera, la discriminación salarial y la falta de visibilidad y reconocimiento.

Además, la cultura académica y científica ha sido históricamente dominada por hombres, lo que significa que muchas de las ideas y premisas subyacentes a estos campos fueron formuladas y promovidas por hombres. Esto ha llevado a un sesgo inconsciente en la evaluación de los méritos científicos y en la selección de los premios.

```python
# Recorte los datos en la zona donde se concentran más los puntos
# esto es en el eje X: entre 0 y 100
# y en el eje Y: entre 0 y 4

filtered_df = dfnobel1[(dfnobel1['edad_cuando_gano'] <= 100) & (dfnobel1['compartido_con'] <= 4)]

# Pinte en colores los puntos por debajo y por encima de la media de Edad cuando Gano.
colors=['coral','teal']

# Otorgue tamaño a los puntos a graficar.
sizes=[25,25]
 
f1 = filtered_df['edad_cuando_gano'].values
f2 = filtered_df['compartido_con'].values
 
# Pinte en colores los puntos por debajo y por encima de la media de Edad cuando Gano.

asignar=[]
for index, row in filtered_df.iterrows():
    if(row['edad_cuando_gano']>60):
        asignar.append(colors[0])
    else:
        asignar.append(colors[1])
        
plt.rcParams["figure.figsize"] = (20, 10)    
plt.scatter(f1, f2, c=asignar, s=sizes[0])

plt.show()
```
